This Computer vision learner (CNN) helps to automatically classify seed plants. All these Spermatophytes are classified by different divisions (see: https://en.wikipedia.org/wiki/Spermatophyte)

The concept of this code is based on lesson 2 of the fast.ai library v3 (see: https://github.com/fastai/course-v3/blob/master/nbs/dl1/lesson2-download.ipynb)

```python
# Import 
from fastai.vision import *
from fastai.widgets import *
```

### 1. Create a directory for classification
The first step is to create directories and a classification.

```python
folder1 = "Cycads"
file1 = "Cycads_url.txt"
```

```python
folder2 = "Conifers"
file2 = "Conifers_url.txt"
```

```python
folder3 = "Flowering_plants"
file3 = "Flowering_plants_url.txt"
```

```python
path = Path('data/plants')
dest1 = path/folder1
dest2 = path/folder2
dest3 = path/folder3
dest1.mkdir(parents=True, exist_ok=True)
dest2.mkdir(parents=True, exist_ok=True)
dest3.mkdir(parents=True, exist_ok=True)
```

### 2. Classify and download images
To search and download the images we used the suggestions of Lesson 2 (Version 3, fast.ai).

```python
classes = ['Cycads','Conifers','Flowering_plants']
```

```python
# Download Cycads
download_images(path/folder1/file1, dest1, max_pics=400)
```

```python
# Download Conifers
download_images(path/folder2/file2, dest2, max_pics=400)
```

```python
# Download Flowering plants
download_images(path/folder3/file3, dest3, max_pics=400)
```

```python
# Check and remove images; see: doc(verify_images)
for c in classes:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
```

### 3. Data augmentation and data bunch
Before we train the model we have to implement data augmentation and create a data bunch.

```python
# Implement data augmentation
tfms = get_transforms(do_flip=True, flip_vert=True, max_zoom=1.1)
```

```python
# Create a data bunch
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=tfms, size=224, num_workers=4).normalize(imagenet_stats)
```

```python
# Look at some pictures
data.show_batch(rows=3, figsize=(12,9))
```

### 4. Train the model
Now we train the resnet34 model with a learner.

```python
# Create a CNN with resnet34
learn = create_cnn(data, models.resnet34, metrics=error_rate)
```

```python
# Learn model
learn.fit_one_cycle(4)
```

```python
# Save model
learn.save('Classifier_stage-1')
```

```python
# Learning rate finder
learn.lr_find()
```

```python
# Plot the learning rate
learn.recorder.plot()
```

```python
# Learn model with a different learning rate
learn.fit_one_cycle(2, 1e-3)
```

```python
# Save model
learn.save('Classifier_stage-2')
```

### 5. Interpret the results
Afterwards, we should interpret our first training.

```python
# Load model
learn.load('Classifier_stage-2')
```

```python
# Classify interpretation
interp = ClassificationInterpretation.from_learner(learn)
```

```python
# Plot a confusion matrix
interp.plot_confusion_matrix()
```

### 6. Cleaning data per human interaction
We can reduce the error rate through classification by hand.

```python
# Implement datasetformatter
ds, idxs = DatasetFormatter().from_toplosses(learn, ds_type=DatasetType.Valid)
```

```python
# Cleaning the data by hand
ImageCleaner(ds, idxs, path)
```

### 7. Implement .csv and train model
Afterwards, we have to implement our new classification schema.

```python
# Create a data bunch
np.random.seed(42)
data = ImageDataBunch.from_csv(path, folder=".", valid_pct=0.2, csv_labels="cleaned.csv",
        ds_tfms=tfms, size=224, num_workers=4).normalize(imagenet_stats)
```

```python
# Create a CNN with resnet34
learn = create_cnn(data, models.resnet34, metrics=error_rate)
```

```python
# Learn model
learn.fit_one_cycle(4)
```

```python
# Save model
learn.save('Classifier_stage-3')
```

```python
# Learning rate finder
learn.lr_find()
```

```python
# Plot the learning rate
learn.recorder.plot()
```

```python
# Learn model
learn.fit_one_cycle(4, 1e-2)
```

```python
# Save model
learn.save('Classifier_stage-4')
```

### 8. Fine-tuning the model
To fine-tune the resnet34 model we unfreeze all layers and learn with 40 epochs and a slicing learning rate.

```python
# Create a CNN with resnet34
learn = create_cnn(data, models.resnet34, metrics=error_rate, ps=0, wd=0)
```

```python
# Unfreeze all layers
learn.unfreeze()
```

```python
# Learn model with 40 epochs and a slicing learning rate
learn.fit_one_cycle(40, slice(1e-4,1e-2))
```

```python
# Save model
learn.save('Classifier_stage-5')
```

```python
# Freeze model
learn.freeze()
```

### 9. Review
Finally, we review our results.

```python
# Classify interpretation
interp = ClassificationInterpretation.from_learner(learn)
```

```python
# Plot a confusion matrix
interp.plot_confusion_matrix()
```

```python
# Plot train and validation losses
learn.recorder.plot_losses()
```

```python
# Plot images
interp.plot_top_losses(9, figsize=(15,11))
```

```python
# Plot the most confusing images
interp.most_confused()
```

```python
# Save the final model
learn.save('Classifier_stage-6')
```
